```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head(1)
```

```python
credit=pd.read_csv("tmdb_5000_credits.csv")
credit.head(1)
```

```python
credit.iloc[0].cast
```

```python
movies= movies.merge(credit, on='title')
movies.head(1)
```

```python
movies.shape
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace= True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']= movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter <3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast']= movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tag']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
movies_new= movies[['movie_id','title','tag']]
movies_new.head()
```

```python
movies_new['tag'] = movies_new['tag'].apply(lambda x: " ".join(x) if isinstance(x, (list, tuple)) else (x if isinstance(x, str) else ""))
movies_new.head()
```

```python
movies_new['tag'][0]
```

```python
movies_new['tag'] =movies_new['tag'].apply(lambda x: x.lower())
movies_new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_new['tag']=movies_new['tag'].apply(stem)
```

```python
movies_new['tag'][0]
```

```python
movies_new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies_new['tag']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = movies_new[movies_new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
